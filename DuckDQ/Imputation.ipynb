```python
import pandas as pd

df = pd.read_json('../Data/stockTicks.json', lines=True)

missing_percentage = (df['symbol'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'symbol': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['date'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'date': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['time'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'time': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['price'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'price': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['quantity'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'quantity': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['buysell'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'buysell': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['ordertype'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'ordertype': {:.2f}%".format(completeness_percentage))

missing_percentage = (df['ipaddr'].isnull().sum() / len(df)) * 100
completeness_percentage = 100.0 - missing_percentage
print("Percentage of completeness in 'ipaddr': {:.2f}%".format(completeness_percentage))
```
